```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
# Model name
model_name = "QIAIUNCC/EYE-Llama_gqa"

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Load model with memory optimization
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,  # Lower precision to save memory
    device_map="auto"           # Auto-manage GPU usage
)
```

```python
# Load pipeline
chatbot_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=200)
```

```python
def ask_question(question):
    prompt = f"User: {question}\nAI:"
    response = chatbot_pipeline(prompt)[0]["generated_text"]

    # Extract response after "AI:"
    return response.split("AI:")[-1].strip()
```

```python
user_question = "What is glaucoma and how is it treated?"
response = ask_question(user_question)

print("Chatbot Response:", response)
```

```python
model.save_pretrained("eye_llama_model")
tokenizer.save_pretrained("eye_llama_model")
```
